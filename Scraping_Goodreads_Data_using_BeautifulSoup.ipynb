###Import Modules

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

###Request page source from URL

```python
url = "https://www.goodreads.com/book/most_read"
```

```python
HEADERS = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}
```

```python
page = requests.get(url, headers=HEADERS)
page
```

```python
#display the page source code
page.content
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
#scraping books title
scraped_books = soup.find_all('a', class_="bookTitle")
scraped_books
```

```python
#parse books names
books = []
for book in scraped_books:
    # Find the span with the title inside the anchor tag
    title_tag = book.find("span", {"itemprop": "name"})
    if title_tag:
        title = title_tag.get_text().strip()
        books.append(title)
books
```

```python
#span class="greyText smallText uitext"
scraped_ratings = soup.find_all('span', class_="minirating")
scraped_ratings
```

```python
# parse ratings
ratings = []
for rating in scraped_ratings:
    text = rating.get_text().replace('\n', '').strip()
    avg_rating = text.split('avg rating')[0].strip()
    ratings.append(avg_rating)
ratings
```

```python
data = pd.DataFrame()
data['Books Names'] = books
data['Ratings'] = ratings
data.head()
```

```python
data.to_csv('Goodreads Top Books.csv', index=False)
```
